```python
import happybase
```

```python
HOSTS = ['mipt-node0{}.atp-fivt.org'.format(i) for i in range(1, 3)]
```

```python
connection = happybase.Connection(HOSTS[1])
```

```python
connection.open()
```

Если таблица существует - удалим ее

```python
connection.disable_table('akhtyamovpavel_wordcount')
connection.delete_table('akhtyamovpavel_wordcount')
```

```python
connection.tables()
```

```python
connection.create_table(
    'akhtyamovpavel_wordcount',
    families={
        'count': dict()
    }
)
```

```python
table = connection.table('akhtyamovpavel_wordcount')
```

```python
for key, data in table.scan(row_prefix=b'10_a'):
    print(key, data)
```

```python
for key, data in table.scan(row_prefix=b'', columns=[b'count:ge_10000000']):
    print(key, data)
```

```python
table.families()
```

# Counters

```python
connection.create_table(
    'akhtyamovpavel_counters',
    families={
        'minute': {
            'time_to_live': 60
        },
        '10s': {
            'time_to_live': 10
        },
        '300s': {
            'time_to_live': 300
        }
    }
)
```

```python
table_counters = connection.table('akhtyamovpavel_counters')
```

```python
table_counters.counter_inc(b'pavel@gmail.com', '10s:click')
```

```python
table_counters.counter_get('pavel@gmail.com', '10s:click')
```
